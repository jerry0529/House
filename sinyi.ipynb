```python
from bs4 import BeautifulSoup
import requests
import json
import pprint
import time
import re
import sqlite3
from datetime import datetime
```

```python
with sqlite3.connect('sinyi.sqlite') as conn:
    c = conn.cursor()
    try:
        c.execute('drop table rentalSinyi')
        print('drop table rentalSinyi')
    except:
        c.execute('''
            create table rentalSinyi(
            url text unique not null, 
            soup text not null, 
            nosql text, 
            ql text,
            getTime text)
        ''')
        print('create table rentalSinyi')
```

```python
n=0
for i in [1,4,5,7,21,26]:
    
    #台北市b=1,桃園市=4,新竹縣=5,新北市=7,新竹市=21,基隆=26
    href = 'http://rent.sinyi.com.tw/AJAX/ajax_searchItem.php?search=1&b={}&page=1'.format(i)
    res = requests.get(href)
    soup = BeautifulSoup(res.text, "lxml")
    
    selPage = soup.select('div.paginate > a.last')[0]['href']
    finalPage = re.findall('[0-9]+',selPage)[0]
    
    for j in range(1,int(finalPage)+1):
        
        href = 'http://rent.sinyi.com.tw/AJAX/ajax_searchItem.php?search=1&b={}&page={}'.format(i, j)
        res = requests.get(href)
        soup = BeautifulSoup(res.text, "lxml")
        selHref = soup.select('ul > li.entry > div.block > div.clearfix > div.lt > a')
        time.sleep(1)
        
        for k in range(0,len(selHref)):
            n += 1
            print(selHref[k]['href'] +','+ str(n))
    time.sleep(1)
    
```

```python
with sqlite3.connect('sinyi.sqlite') as conn:
    c = conn.cursor()
    
    for i in [1,4,5,7,21,26]:

        #台北市b=1,桃園市=4,新竹縣=5,新北市=7,新竹市=21,基隆=26
        href = 'http://rent.sinyi.com.tw/AJAX/ajax_searchItem.php?search=1&b={}&page=1'.format(i)
        res = requests.get(href)
        soup = BeautifulSoup(res.text, "lxml")

        selPage = soup.select('div.paginate > a.last')[0]['href']
        finalPage = re.findall('[0-9]+',selPage)[0]

        for j in range(1,int(finalPage)+1):

            href = 'http://rent.sinyi.com.tw/AJAX/ajax_searchItem.php?search=1&b={}&page={}'.format(i, j)
            res = requests.get(href)
            soup = BeautifulSoup(res.text, "lxml")
            selHref = soup.select('ul > li.entry > div.block > div.clearfix > div.lt > a')
            #time.sleep(2)

            for k in range(0,len(selHref)):

                selhref = selHref[k]['href']
                selres = requests.get(selhref)
                selsoup = BeautifulSoup(selres.text, 'lxml')
                try:
                    selselect = selsoup.select('#itemDetail > div.main')[0].text
                except:
                    print('lxml not found')
                getTimes = str(datetime.now())
                
                try:
                    insert_string = "insert or ignore into rentalSinyi (url, soup, getTime) VALUES (?, ?, ?)"
                    c.execute(insert_string, (selhref, selselect, getTimes))
                    conn.commit()
                
                except Exception as e:
                    print('Somthing wrong at :' + selhref + str(e))
                    
                #print(selhref, selselect, getTimes)

        #time.sleep(2)
```
